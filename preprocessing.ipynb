```python
import os
import re
import glob, random
import cv2
import numpy as np
from tqdm import tqdm_notebook
import matplotlib.pyplot as plt
```

```python
##read frames from a video and load them into col_images2

myFrameNumber = 50
cap = cv2.VideoCapture('blink_detection_demo.mp4')

# get total number of frames
totalFrames = cap.get(cv2.CAP_PROP_FRAME_COUNT)
print("number of frames: {}".format(totalFrames))
col_images=[]
while cap.isOpened():
    ret, frame = cap.read()
    # print(ret)
    if (ret):
      frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
      col_images.append(frame)
    else:
      break
    if cv2.waitKey(20) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
